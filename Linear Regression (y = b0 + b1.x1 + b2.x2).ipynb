```python
x = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y = [1, 2, 2.5, 2.5, 3.3, 3.5, 4, 3.75, 5]
z = [1, 3, 5, 3, 2, 1, 6, 4, 5]
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)
plt.show()
```

```python
b0 = 0.7
b1 = 0.8
b2 = 0.9
alpha = 0.0001


for j in range(100000):
    for i in range(len(x)):
        x1, x2, ya = x[i], y[i], z[i]
        yp = b0 + b1*x1 + b2*x2
        b0 = b0 + alpha*(ya-yp)
        b1 = b1 + alpha*(ya-yp)*x1
        b2 = b2 + alpha*(ya-yp)*x2 

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection = '3d')

X = np.arange(0, 7, 0.25)
Y = np.arange(0, 7, 0.25)
X, Y = np.meshgrid(X, Y)
Z = b0 + b1*X + b2*Y

surf = ax.plot_surface(X, Y, Z, alpha=0.1)
ax = plt.gca()
ax.hold(True)

ax.scatter(x, y, z, color="red")


plt.show()
print b0, b1, b2
```
