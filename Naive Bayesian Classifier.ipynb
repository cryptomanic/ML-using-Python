```python
import pandas as pd
import numpy as np
from itertools import compress

df = pd.read_csv("/home/hp/Desktop/ML-using-Python/tennis.csv")
df
```

```python
def ProbXGivenC(X, Class):
    dfc = df.loc[df.play == Class]
    A = dfc.columns.values
    ans = 1.0
    for i in range(len(A)-1):
        ans = ans * sum(dfc[A[i]] == X[i]) / len(dfc)
    return ans

def predict_class(X, C, PC):
    '''
        X : input vector
        C : probability distribution of classes
        PC : possible classes
        
        e.g.
        X : ["rainy", "hot", "high", True]
        C : [0.5, 0.5]
        PC : ['yes', 'no']
    '''
    N = len(PC)
    prob = [0 for i in range(N)]
    for i in range(N):        
        prob[i] = C[i]*ProbXGivenC(X, PC[i])
            
    # find and return the class
    return list(compress(PC, np.array(prob) == max(prob)))

print predict_class(["rainy", "hot", "high", True], [0.5, 0.5], ['yes', 'no'])
print predict_class(["sunny", "hot", "high", False], [0.5, 0.5], ['yes', 'no'])
print predict_class(["overcast", "hot", "normal", False], [0.5, 0.5], ['yes', 'no'])
```

```python
# check training error
train_res = [predict_class([df.outlook[i], df.temp[i], df.humidity[i], df.windy[i]], [9.0/14, 5.0/14], ['yes', 'no'])[0] for i in range(len(df))]
num_correct = sum(np.array(train_res) == np.array(df.play))
print num_correct, 'out of 14 are correctly predicted on training data'
```
